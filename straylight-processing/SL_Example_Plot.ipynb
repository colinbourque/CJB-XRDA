```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
from xsutils import sl_qdp
```

```python
a,b = sl_qdp('sl_data/40201001002A-all.qdp')

fig, (ax1, ax2) = plt.subplots(2,1,figsize=(8,7), sharex=True, height_ratios=[2,1])
fig.subplots_adjust(hspace=0)

## setup
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlim([3,60])
ax1.set_ylim([5e-5, .2])
ax1.set_ylabel('counts s$^{-1}$ keV$^{-1}$')
ax1.set_title('40201001002A')

## ldata
ax1.errorbar(x=a[0], y=a[2],
           xerr=a[1], yerr=a[3],
           color='k', ls='none', lw=.8)

ax1.plot(a[0], a[4], color='darkgrey', label='Total Model', lw=.5)
ax1.plot(a[0], a[5], color='b', label='Source Model', lw=.5)
ax1.plot(a[0], a[10], color='b', ls='--', lw=.5)
ax1.plot(a[0], a[11], color='b', ls='--', lw=.5)
ax1.plot(a[0], a[12], color='b', ls='--', lw=.5)

## Use this line to plot the total background
ax1.plot(a[0], (a[6]+a[7]+a[8]+a[9]), color='r', ls='-.', label='Total background', lw=.5)

## Use these lines to plot each background component individually
# ax1.plot(a[0], a[6], color='g', ls='-', label='fCXB', lw=.5)
# ax1.plot(a[0], a[7], color='r', ls='-', label='Instrument bgd', lw=.5)
# ax1.plot(a[0], a[8], color='purple', ls='-', label='aCXB', lw=.5)
# ax1.plot(a[0], a[9], color='y', ls='-', label='Constant bgd', lw=.5)

ax1.legend()

## delchi
ax2.set_xscale('log')
ax2.set_yscale('linear')
ax2.set_xlim([3,120])
ax2.set_xlabel('Energy (keV)')
ax2.set_ylabel('(data-model)/error')

ax2.errorbar(x=b[0], y=b[2],
           xerr=b[1], yerr=b[3],
           color='k', ls='none', lw=.8)

ax2.axhline(y=0, color='tab:blue')

## x-axis energy labels
ax2.xaxis.set_major_formatter(matplotlib.ticker.ScalarFormatter())
ax2.set_xticks([3,6,10,30,60,100])
ax2.set_ylim([-5,5])

for ax in [ax1,ax2]:
    ax.tick_params(axis='both', direction='in', which='major')
    ax.tick_params(axis='both', direction='in', which='minor')
    ax.xaxis.set_ticks_position('both')
    ax.yaxis.set_ticks_position('both')

# plt.savefig('nu40201001002A-spectrum.eps', bbox_inches='tight')
plt.show()
```
